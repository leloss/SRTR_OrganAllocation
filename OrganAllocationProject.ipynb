##Analysis of DONOR_DISPOSITION

```python
# Format Name	Code	Descript
# FDBDSP	1	1: Consent Not Requested
# FDBDSP	2	2: Consent Not Obtained
# FDBDSP	3	3: Organ Not Recovered
# FDBDSP	4	4: Recovered Not for Tx
# FDBDSP	5	5: Recovered for TX but Not Tx
# FDBDSP	6	6: Transplanted
# FDBDSP	7	7: N/A
```

```python
e=db.cursor()
e.execute("""select DONOR_ID, DON_ORG, DON_DISPOSITION, DON_REASON_CD, DON_DISCARD_CD, PX_ID from donor_disposition where (DON_ORG='LKI' or DON_ORG='RKI' or DON_ORG='EKI') and DON_DISPOSITION=5 order by DONOR_ID""")
notTransped = e.fetchall()
e.execute("""select DONOR_ID, DON_ORG, DON_DISPOSITION, DON_REASON_CD, DON_DISCARD_CD, PX_ID from donor_disposition where (DON_ORG='LKI' or DON_ORG='RKI' or DON_ORG='EKI') and DON_DISPOSITION=6 order by DONOR_ID""")
Transped = e.fetchall()
```

```python
import numpy as np
# notTransped = np.unique(notTransped)

transp_list=[]
transp_reason_list=[]
nottransp_list=[]
nottransp_reason_list=[]
nottransp_disc_list=[]
for row in notTransped:
    DONOR_ID = row[0]
    PTR_OFFER_ACPT = row[1]
    DON_REASON_CD = row[3]
    DON_DISCARD_CD = row[4]
    nottransp_list.append(int(DONOR_ID))
    nottransp_reason_list.append(DON_REASON_CD)
    nottransp_disc_list.append(DON_DISCARD_CD)
    
# Transped = np.unique(Transped)
for row in Transped:
    DONOR_ID = row[0]
    PTR_OFFER_ACPT = row[1]
    DON_REASON_CD = row[3]
    DON_DISCARD_CD = row[4]
    transp_list.append(int(DONOR_ID))
    transp_reason_list.append(DON_REASON_CD)
    
print len(np.unique(nottransp_list)), len(np.unique(transp_list)), float(len(np.unique(nottransp_list)))/float(len(np.unique(transp_list)))
```

```python
import csv

table_discardcodes={}
with open('./SRTR_DiscardCodes.csv', mode='r') as infile:
    reader = csv.reader(infile)
    table_discardcodes={int(rows[1]):rows[2] for rows in reader}
```

```python
import csv

table_reasoncodes={}
with open('./SRTR_ReasonCodes.csv', mode='r') as infile:
    reader = csv.reader(infile)
    table_reasoncodes={int(rows[1]):rows[2] for rows in reader}
```

###Not Transplanted

```python
uniq_nottransp_disc=np.unique(nottransp_disc_list)
data_nottransp=[]
for x in uniq_nottransp_disc:
    if x==None:
        print x, nottransp_disc_list.count(x), 100*nottransp_disc_list.count(x)/float(len(nottransp_disc_list)),'%'
        data_nottransp.append([100*nottransp_disc_list.count(x)/float(len(nottransp_disc_list)), 'None: Not Known'])
    else:        
        print x, nottransp_disc_list.count(x), 100*nottransp_disc_list.count(x)/float(len(nottransp_disc_list)),'%', table_discardcodes[int(x)]
        data_nottransp.append([100*nottransp_disc_list.count(x)/float(len(nottransp_disc_list)), table_discardcodes[int(x)]])
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 6))
N = len(data_nottransp)
Y = [x[0] for x in data_nottransp] 
# print Y
ind = np.arange(N)   
# print N, ind                # the x locations for the groups
width = 1       
plt.bar(ind, Y, width)
# print rects

# axes and labels


plt.xlim(-width,len(ind)+width)
# plt.ylim(0,45)
# plt.ylabel('Scores')
# ax.set_title('Scores by group and gender')
xTickMarks = [str(x[1]) for x in data_nottransp]
# plt.xticks(ind+width)
# plt.xticks([-1, 0, 1, 2], ['NA', 'Homozygote 1/1', 'Heterozygote','Homozygote 2/2'])
plt.xticks(ind+.5,xTickMarks, rotation=90, fontsize=12)
# plt.setp(xtickNames, rotation=45, fontsize=10)
plt.ylabel('p(x) (%)')
plt.grid('on')
plt.title('Total Organs Not Transplanted = '+str(len(np.unique(nottransp_list))))
plt.show()
```
